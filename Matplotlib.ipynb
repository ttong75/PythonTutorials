```python
plants = [ "Spinach", "Turnip", "Rhubarb", "Broccoli", "Kale"]
died = [10,25,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize = (6,6))
```

```python
plt.bar(plants, germinated, width = 0.5, color = "orchid")
plt.title("Germination Rates")
plt.ylabel("Number of Seedlings")
```

```python
plt.bar(plants, died, width = 0.5, color = "red")
plt.title("Failure Rates")
plt.ylabel("Number of Dead Seedlings")
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants,died, color = "red")
plt.bar(plants, germinated, width = 0.4, color = "pink")
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants,died, color = "red")
plt.bar(plants, germinated, width = 0.4, color = "pink", align = "edge")
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize = (6,6))
plt.bar(plants,died, color = "red", bottom = germinated, label = "Failed Seedlings")
plt.bar(plants, germinated, color = "pink", label = "Germianted Seedlings")
plt.legend(shadow = True, frameon = True, facecolor = "white")
plt.title("Greenhouse Seedlings")
```

* bottom - stacks the data - sums it together

```python
plt.figure(figsize = (6,6))
plt.barh(plants,died, color = "red")
```

```python
import pandas as pd
titanic = pd.read_csv("data/titanic.csv")
```

```python
titanic["age"] = pd.to_numeric(titanic["age"], errors = "coerce")
```

```python
plt.hist(titanic["age"])
```

```python
first_class = titanic[titanic["pclass"] == 1]["age"]
second_class = titanic[titanic["pclass"] == 2]["age"]
third_class = titanic[titanic["pclass"] == 3]["age"]
```

```python
plt.hist(first_class)
```

```python
plt.hist(first_class, label = "1st Class", alpha = 0.5, bins = 20)
plt.hist(second_class, label = "2nd Class", alpha = 0.5, bins = 20)
plt.hist(third_class, label = "3rd Class", alpha = 0.5, bins = 20)
plt.legend()
plt.title("Titanic Passengers")
plt.xlabel("Passenger Age")
```

* plt. subplot (number of rows, columns, which index/axes currently working on

```python
nums = np.arange(5)
plt.figure(figsize = (7,4))
plt.suptitle ("First Subplot", fontsize = 20)
plt.subplot(1,3,1)
plt.plot(nums,nums)
plt.title("X")

plt.subplot(1,3,2)
plt.plot(nums**2, color = "orange")
plt.title("X Squared")

plt.subplot(1,3,3)
plt.plot(nums**3, color = "red")
plt.title("X Cubed")
plt.tight_layout()
```

### * Order Matters!

```python
plt.figure(figsize = (10,4))
plt.suptitle("Titanic Passengers' Ages")

plt.subplot(1,3,1)
plt.hist(first_class, label = "1st Class", bins = 20, color = "red")
plt.title("1st Class")

plt.subplot(1,3,2)
plt.hist(second_class, label = "2nd Class", bins = 20, color = "yellow")
plt.title("2nd Class")

plt.subplot(1,3,3)
plt.hist(third_class, label = "3rd Class", bins = 20)
plt.title("3rd Class")
plt.show()
plt.tight_layout
```

* if you want the same y values:

```python
plt.figure(figsize = (10,4))

plt.subplot(1,3,1)
plt.hist(first_class, label = "1st Class", bins = 20, color = "red")
plt.title("1st Class")
ax = plt.subplot(1,3,1)

plt.subplot(1,3,2, sharey=ax)
plt.hist(second_class, label = "2nd Class", bins = 20, color = "yellow")
plt.title("2nd Class")

plt.subplot(1,3,3, sharey = ax)
plt.hist(third_class, label = "3rd Class", bins = 20)
plt.title("3rd Class")
plt.show()
plt.tight_layout
```
