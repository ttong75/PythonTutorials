```python
df = titanic.head()
women = df.sex == "female"
df[~ women]
```

* ~ (tilda) gives you the opposite of what you typed - Bitwise negation

```python
newly_built = houses["yr_built"]>=2014
newly_renovated = houses["yr_renovated"] >=2014
recent_homes = newly_built | newly_renovated
houses[recent_homes]
```

```python
houses[~recent_homes]
```
